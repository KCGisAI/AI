```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("google/gemma-2b")
model = AutoModelForCausalLM.from_pretrained("google/gemma-2b", device_map="auto")
```

```python
from datasets import load_dataset

data = load_dataset("fancyzhx/ag_news")
data['train']
```

```python
def preprocess_function(examples):
  text = examples['text']
  return tokenizer(text, max_length = 200, truncation=True)

tokenized_data = data.map(preprocess_function, batched = True)
tokenized_data
```

```python
import numpy as np
from tqdm import tqdm

import torch

def zero_shot_classification(text, task_description, labels):  # text는 주어진 입력, task_description은 task에 대한 설명, labels은 class들을 text로 변환한 결과입니다.
    text_ids = tokenizer(task_description + text, return_tensors="pt").to("cuda")  # 먼저 task_description과 text를 이어붙인 후, tokenize합니다.
    probs = []
    for label in labels:  # 그 다음 각 text화된 label들을 tokenize하고 입력에 이어붙인 후, Gemma-2B에 넣어줍니다.
        label_ids = tokenizer(label, return_tensors="pt").to("cuda")
        n_label_tokens = label_ids['input_ids'].shape[-1] - 1  # text로 변환한 label의 token 수를 계산합니다
        #레이블의 첫 번째 토큰(보통 시작 토큰)을 제외하고 연결

        input_ids = {
            'input_ids': torch.concatenate([text_ids['input_ids'], label_ids['input_ids'][:, 1:]], axis=-1),  # concatenate 명령어를 통해 이어붙이는 모습입니다.
            'attention_mask': torch.concatenate([text_ids['attention_mask'], label_ids['attention_mask'][:, 1:]], axis=-1)
        }

        logits = model(**input_ids).logits  # Logit을 계산한 모습입니다.
        prob = 0
        n_total = input_ids['input_ids'].shape[-1]
        for i in range(n_label_tokens, 0, -1):  # 일반적으로 text로 변환한 label은 여러 token으로 이루어져있습니다. 이러한 label에 대한 logit은 구성하는 모든 token들의 logit들의 합으로 정의합니다.
            token = label_ids['input_ids'][0, i].item()
            prob += logits[0, n_total - i, token].item()
        probs.append(prob)

        del input_ids
        del logits
        torch.cuda.empty_cache()  # 위의 del과 empty_cache() 명령어를 통해 GPU를 제때 할당해제 해줍니다. 만약 GPU가 여유롭다면 지워주시는게 속도적으로 이득입니다.

    return probs
```

```python
import numpy as np

id2label = {
    0 : "World",
    1 : "Sports",
    2 : "Business",
    3 : "Science and Technology"
}

n_corrects = 0
for i in (range(50)):
    text = tokenized_data['test'][i]['text']
    label = tokenized_data['test'][i]['label']
    probs = zero_shot_classification(
        text,
        "Determine the topic of the given sentence. The topics are defined as follows."
        "World Covers global events, international relations, and significant news affecting multiple countries."
        "Sports Includes topics about various sports, game results, athlete performances, and sports culture."
        "Business Relates to market trends, company news, economic issues, and financial reports."
        "Science and Technology Discuss scientific discoveries, technological advancements, research findings, and their implications."
        "Choose the most appropriate topic based on the content of the sentence.",
        labels=["Answer: World", "Answer: Sports", "Answer: Business" , "Answer: Science and Technology"]
    )

    pred = np.argmax(np.array(probs))

    if pred == label:
        n_corrects += 1
    print(f"predicition is {id2label[pred]} and label is {id2label[label]}")
    print(f"accuracy is {100*n_corrects/50}% and n_corrects are {n_corrects}")
```
